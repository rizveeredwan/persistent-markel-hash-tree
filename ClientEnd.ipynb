```python
def do_encryption(raw_blocks):
    encrypted_blocks = []
    for i in raw_blocks:
      encrypted_blocks.append(security_implementation.encrypt_block(i))
    return encrypted_blocks
```

```python
def text_encryption(text_file):
  # File encryption at user end
  # getting raw blocks
  raw_blocks = security_implementation.padding_64(text_file, chunk_size=security_implementation.BLOCK_SIZE)
  encrypted_blocks = do_encryption(raw_blocks)
  # print("done")
  # print(encrypted_blocks)
  return encrypted_blocks
```

```python
def do_decryption(encrypted_blocks):
  decrypted_blocks = []
  for i in range(0, len(encrypted_blocks)):
    decrypted_blocks.append(security_implementation.decrypt_block(encrypted_blocks[i]))
  return decrypted_blocks
```

```python
class Client:
  def __init__(self):
    self.tpa_object = tpa.TPA(server_segment_tree_control=server_segment_tree_control, security_implementation=security_implementation)
    self.version_root = [] 

  def send_file_to_server(self, text_file):
    encrypted_blocks = text_encryption(text_file)
    temp = server_segment_tree_control.initial_file_upload(encrypted_blocks)
    self.version_root.append(temp)
    # print(self.version_root[0])
    # keeping the root hash 
    self.tpa_object.add_new_version_root_hash(new_version_root_hash=self.version_root[-1].data)
    # save the base encrypted blocks hash 
    self.tpa_object.initial_file_block_save(encrypted_blocks_hash=security_implementation.calcFileBlockHashes(encrypted_blocks))
    return encrypted_blocks

  def update_file_to_server(self, block_text, block_to_update, bruteforce=False):
    #print(len(block_text))
    block_text = block_text.ljust(security_implementation.BLOCK_SIZE,"#")
    encrypted_block = security_implementation.encrypt_block(block_text)
    #print(len(encrypted_block), len(block_text))
    temp = server_segment_tree_control.update_existing_file(encrypted_block=encrypted_block, block_to_update=block_to_update, bruteforce=bruteforce)
    # adding updated root for new version 
    self.version_root.append(temp) 
    # adding root hash to TPA
    self.tpa_object.add_new_version_root_hash(new_version_root_hash=self.version_root[-1].data)
    # store the updated block hash: saving the encryption block hash
    # print(encrypted_block, security_implementation.calcBlockHash(encrypted_block.encode('utf-8')))
    self.tpa_object.store_updated_blocks(updated_encryption_block_hash=security_implementation.calcBlockHash(encrypted_block.encode('utf-8')), block_id=block_to_update)
    return 

  def request_for_file_retrival(self, file_version_no):
    encrypted_blocks, tree_root_hash = server_segment_tree_control.file_retrival_by_client(file_version_no)
    # print(encrypted_blocks)
    """
    # debug
    for i in range(0, len(encrypted_blocks)):
      print(encrypted_blocks[i], len(encrypted_blocks[i]), len(encrypted_blocks[i])%16)
    """
    if(tree_root_hash == self.version_root[file_version_no].data):
      # get decrypted blocks 
      decrypted_blocks = do_decryption(encrypted_blocks)
      # assertion condition: if all the blocks appeared 
      assert(len(decrypted_blocks) == len(self.tpa_object.base_encrypted_blocks_hash))
      return decrypted_blocks
    else:
      print("File version root hash did not match")
      return []

  def challenge_server(self, encrypted_blocks, file_version_no, block_index, hashes=True):
    sibling_nodes = server_segment_tree_control.challenge_server(file_version_no, block_index, hashes)
    if hashes == True:
      node_based = False # only hashes are kept 
    else:
      node_based = True # nodes are taken 

    hash = security_implementation.calculate_root_from_sib_path(input_hash=security_implementation.calcBlockHash(encrypted_blocks[block_index].encode('utf-8')), 
                                      sibling_nodes=sibling_nodes, node_based=node_based)
    return (hash == self.version_root[file_version_no].data)

  def server_seg_tree_memory_calc(self):
    # memory used 
    """
    process = psutil.Process(os.getpid())
    _list.append(process.memory_info().rss/(1024.0*1024.0)) # in bytes -> MB
    """
    server_segment_tree_control._TOTAL_SIZE_IN_BYTES = 0.0 
    _dict={}
    array = []
    prev = 0
    _sum = 0
    total_cum_memory = []
    for i in range(0, len(self.version_root)):
      server_segment_tree_control.recursive_size_check(node=self.version_root[i], _dict=_dict)
      array.append(server_segment_tree_control._TOTAL_SIZE_IN_BYTES - prev) 
      total_cum_memory.append(server_segment_tree_control._TOTAL_SIZE_IN_BYTES)
      prev = server_segment_tree_control._TOTAL_SIZE_IN_BYTES
      if(i>0):
        _sum = _sum + array[-1] 
    if(_sum > 0):
      _sum = (_sum*1.0) /(len(self.version_root)-1)
    return server_segment_tree_control._TOTAL_SIZE_IN_BYTES, _sum, array[0],total_cum_memory  # total memory usage, avg memory usage, initial memory usage, total memory usage points 
```

```python
def get_file_size(file_name):
  value = 0
  for i in range(0, len(file_name)):
    if file_name[i] >= '0' and file_name[i] <='9':
      value = value * 10 + int(file_name[i])
    else:
      break 
  return value  
```

```python
class FileBasedStat:
  def __init__(self, number_of_files, runner_flag="rizvee"):
    self.number_of_files = number_of_files
    if runner_flag.lower() == "rizvee":
      self.path = os.path.join('/content') 
    else:
      self.path = os.path.join('/content') 

  def random_generation_text(self, char_number):
    _str = ""
    for i in range(0, char_number):
      _str = _str + chr(random.randint(0,25)+ord('a')) 
    return _str 

  def initiate(self,  bruteforce=False, stat_file_name='stat.csv', memory_change_stat='5MB_memory_change_stat.csv', num_updates=1000):
    global client
    # bruteforce means many separate segment trees 
    new_path = os.path.join(self.path, stat_file_name)
    new_path2 = os.path.join(self.path, memory_change_stat)
    w = open(new_path, 'w', encoding='utf-8')
    w2 = open(new_path2, 'w', encoding='utf-8')
    writer = csv.writer(w)
    writer2 = csv.writer(w2)
    writer.writerow(['file_name', 'initial_file_upload(sec.)', 'avg_block_update_time(sec.)',  'avg_challenge_server_time(sec.)',  'avg_file_retrival(sec.)', 
                     'Total Memory(MB)', 'avg. update memory (MB)', 'initial memory (MB)'])
    files = os.listdir(self.path)
    for j in range(0, len(files)):
      num = get_file_size(file_name=files[j])
      #print(files[j], num, re.search(files[j], "[0-9]+"))
      if re.search("[0-9]+mb\.txt", files[j]) is None:
        continue
      if num == 0: # not an input file 
        continue 
      """
      if num != 128: # only stat for 136 MB
        continue
      """
      file_path = os.path.join(self.path, files[j])
      with open(file_path, 'r') as f:
        print("file_path = ", file_path)
        _list = [] 
        _list.append(files[j])
        lines = f.readlines() 
        print("length of lines = ", len(lines))
        client = Client()
        # initial file upload time 
        start = timer()
        client.send_file_to_server(text_file=lines[0])
        end = timer() 
        _list.append(end-start)
        print(end-start)

        # block update time 
        number_of_updates = num_updates # random.randint(1000,1000) 
        block = random.randint(0,7) 
        _sum = 0 
        for i in range(1, number_of_updates+1):
          #upperbound = max(0, min(100, len(client.tpa_object.base_encrypted_blocks_hash)-block-1))
          #block = block + random.randint(0,upperbound) 
          block = random.randint(0, len(client.tpa_object.base_encrypted_blocks_hash)-1)
          block_text = self.random_generation_text(char_number=random.randint(10,30))
          start = timer()
          client.update_file_to_server(block_text=block_text, block_to_update=block,  bruteforce=bruteforce)
          end = timer()
          _sum = _sum + end-start 
        _sum = _sum / (number_of_updates*1.0) 
        _list.append(_sum)  
        print("DONE")

        # challenge server 
        start = timer()
        number_of_queries = random.randint(5,5)
        modular_cal = client.tpa_object.challenge_server(number_of_queries=number_of_queries)    
        end = timer()  
        _sum = (end - start)/(number_of_queries*1.0)
        print("modular_cal ",modular_cal, _sum)
        _list.append(_sum)  
 
        # file retrival 
        _sum = 0 
        for i in range(1, number_of_updates+1):
          start = timer()
          blocks = client.request_for_file_retrival(file_version_no=i-1)
          end = timer()
          _sum = _sum + end-start 
          print(end-start, len(blocks))
        _sum = _sum /(number_of_updates*1.0)
        _list.append(_sum)

        # memory calc
        total_mem, avg_update_mem, initial_mem, total_cum_memory = client.server_seg_tree_memory_calc()
        total_mem = total_mem/(1024.0 * 1024.0)
        avg_update_mem =  avg_update_mem / (1024.0 * 1024.0)
        initial_mem = initial_mem / (1024.0 * 1024.0)
        _list.append(total_mem)
        _list.append(avg_update_mem)
        _list.append(initial_mem) 
        if bruteforce is True:
          while len(total_cum_memory) < 1000:
            total_cum_memory.append(total_cum_memory[-1]+total_cum_memory[0])
        for i in range(0, len(total_cum_memory)):
          writer2.writerow([i, total_cum_memory[i]/ (1024.0 * 1024.0)])
        del client

        # file writing
        writer.writerow(_list)

        # Returns the number of
        # objects it has collected
        # and deallocated
        collected = gc.collect()
        
        # Prints Garbage collector
        # as 0 object
        print("Garbage collector: collected", "%d objects." % collected)
```
