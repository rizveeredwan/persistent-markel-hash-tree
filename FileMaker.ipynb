```python
class CreateFile:
  def __init__(self):
    pass 
  def create(self, file_name, size_in_kb): 
    f = open(file_name, 'w')
    number_of_bytes = 1000 * size_in_kb
    for i in range(0, number_of_bytes):
      char = chr(random.randint(0,25)+ord('a'))
      f.write(char)
      # print(char) 
```

```python
cr = CreateFile()
for i in range(11,256, 25):
  path = os.path.join('/content','gdrive', 'MyDrive', 'Research', 'persistent-markel-hash-tree', 'custom_data_files')
  path = os.path.join(path, str(i)+"mb.txt")
  cr.create(file_name=path, size_in_kb=1000*i)
```
